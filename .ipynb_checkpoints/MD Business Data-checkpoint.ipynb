```python
def yelp_call(url_param):
    response = requests.get(url, headers=headers, params=url_params)
    return response.json()['businesses']
```

```python

business = data['businesses'][0]
```

```python
biz_tuple = (business['id'],
             business['name'], 
             business['review_count'], 
             business['categories'], 
             business['rating'], 
             business['price'], 
             business['location']['display_address'])
```

```python
def parse_data(list_of_data):
    # create a container to hold our parsed data
    businesses = []
    # loop through our business and 
    for business in list_of_data:
    # parse each individual business into a tuple
        biz_tuple = (business['id'],
                     business['name'], 
             business['review_count'], 
            [x['title'] for x in business['categories']], 
             business['rating'], 
             business['price'], 
             business['location']['display_address'])
    # add each individual business tuple to our data container
        businesses.append(biz_tuple)
    # return the container with all of the parsed results
    return businesses
```

```python
parsed_results = parse_data(data['businesses'])
```

```python
import pandas as pd
```

```python

# create a variable  to keep track of which result you are in. 
cur = 0
total_data = []
#set up a while loop to go through and grab the result 
while cur < 1000:
    #set the offset parameter to be where you currently are in the results 
    url_params['offset'] = cur
    #make your API call with the new offset number
    results = yelp_call(url_params)

    #after you get your results you can now use your function to parse those results
    parsed_results = parse_data(results)
    total_data.extend(parsed_results)
    
    # use your function to insert your parsed results into the db
    #df_save(parsed_results)
    #increment the counter by 50 to move on to the next results
    cur += 50
```

```python
biz_df = pd.DataFrame(total_data, columns = ['id', 'Name', 'Review_Count', 'Categories', 'Rating', 'Price', 'Address'])
```
