```python
va_biz = pd.read_csv('va_biz_reviews')
```

```python
va_biz[va_biz['Review_Count']  <3]
```

```python
total_data = []
for business in va_biz.loc[:,'id']:
    url = 'https://api.yelp.com/v3/businesses/{}/reviews'.format(business)
    response = requests.get(url, headers=headers)
    data = response.json()
    try:
        rev0 = data['reviews'][0]['text']
        rat0 = data['reviews'][0]['rating']
        time0 = data['reviews'][0]['time_created']
        rev1 = data['reviews'][1]['text']
        rat1 = data['reviews'][1]['rating']
        time1 = data['reviews'][1]['time_created']
        rev2 = data['reviews'][2]['text']
        rat2 = data['reviews'][2]['rating']
        time2 = data['reviews'][2]['time_created']
    except:
        rev1 = None
        rat1 = None
        rev2 = None
        rat2 = None
        rev0 = None
        rat0 = None
        time0 = None
        time1 = None
        time2 = None
    biz_tuple =(business,
               rat0,
               rev0,
               time0,
               rat1,
               rev1,
               time1,
               rat2,
               rev2,
               time2)
    total_data.append(biz_tuple)
    
    
```

```python
va_reviews_df = pd.DataFrame(total_data, columns = ['id','Rating_1', 'Review_1', 'Time_Created_1', 'Rating_2', 'Review_2', 'Time_Created_2', 'Rating_3', 'Review_3','Time_Created_3'])
```

```python
va_reviews_df.to_csv('va_reviews')
```
