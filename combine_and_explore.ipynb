```python
df['Categories'].apply(lambda x: list(x))
df['Categories'][3][0]
```

```python
#add sub categories columns
df['cat_1'] = df['Categories'].apply(lambda x: list(x)[0])
df['cat_2'] = df['Categories'].apply(lambda x: list(x)[1] if len(x) >1 else None) 
df['cat_3'] = df['Categories'].apply(lambda x: list(x)[2] if len(x) > 2 else None) 
df.head()
```

```python
#check to see max length of category list
df['Categories'].apply(lambda x: len(x)).max()
```

```python
df['cat_3'].value_counts()
```

```python
#Create new dataframe with categories
cat_3_counts = pd.DataFrame(df['cat_3'].value_counts())
cat_3_counts
```

```python
cat_2_counts = pd.DataFrame(df['cat_2'].value_counts())
```

```python
cat_1_counts = pd.DataFrame(df['cat_1'].value_counts())
```

```python
#Join 3 category Dataframe
cat_1_2 = cat_1_counts.join(cat_2_counts, how = 'outer')
cat_counts_df = cat_1_2.join(cat_3_counts, how = 'outer')
cat_counts_df
```

```python
#Replace NaN with 0
cat_counts_df.fillna(0, inplace = True)
```

```python
#Create Total Column
cat_counts_df['total'] = cat_counts_df['cat_1'] + cat_counts_df['cat_2'] + cat_counts_df['cat_3']
```

```python
#Sort total column in descending order
cat_counts_df.sort_values('total', ascending = False, inplace = True)
```

```python
#Create bar graph of Top 20 Categories
fig,ax = plt.subplots(figsize = (20,10))
x = cat_counts_df.reset_index()['index'][0:20]
y = cat_counts_df['total'][0:20]
ax.barh(x,y)
ax.set_title('Top 20 Categories', fontsize = 15)
ax.set_xlabel('Total', fontsize = 15)
ax.set_ylabel('Categories', fontsize = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
```

```python
#Create dataframe with 
cat_1_values = df.groupby('cat_1').agg({'Review_Count':'sum', 'Rating': 'sum'})
cat_2_values = df.groupby('cat_2').agg({'Review_Count':'sum', 'Rating': 'sum'})
cat_3_values = df.groupby('cat_3').agg({'Review_Count':'sum', 'Rating': 'sum'})
```

```python
cat_1_2_values = cat_1_values.join(cat_2_values, how = 'outer', lsuffix = '_one', rsuffix = '_two')
cat_revcount_rating = cat_1_2_values.join(cat_3_values, how = 'outer')
```

```python
cat_revcount_rating.fillna(0, inplace = True)
cat_revcount_rating.rename(columns = {'Review_Count' : 'Review_Count_three', 'Rating' : 'Rating_three'}, inplace = True)
```

```python
cat_revcount_rating['total_review_count'] = cat_revcount_rating['Review_Count_one'] + cat_revcount_rating['Review_Count_two'] + cat_revcount_rating['Review_Count_three']
```

```python
cat_revcount_rating['total_rating'] = cat_revcount_rating['Rating_one'] + cat_revcount_rating['Rating_two'] + cat_revcount_rating['Rating_three']
```

```python
cat_revcount_rating
```

```python
cat_info_df = cat_counts_df.join(cat_revcount_rating)
cat_info_df.head(2)
```

```python
cat_info_df['avg_review_count'] = cat_info_df['total_review_count'] / cat_info_df['total']
cat_info_df['avg_rating'] = cat_info_df['total_rating'] / cat_info_df['total']
```

```python
cat_info_df
```

```python
#Create bar graph of Top 20 Categories w Review Counts
fig,ax = plt.subplots(figsize = (20,10))
x = cat_info_df.reset_index()['index'][0:20]
y = cat_info_df['avg_review_count'][0:20]
ax.barh(x,y)
ax.set_title('Top 20 Categories By Review Counts', fontsize = 15)
ax.set_xlabel('Average Review Count', fontsize = 15)
ax.set_ylabel('Categories', fontsize = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
```

```python
#Create bar graph of Top 20 Average Review Counts

fig,ax = plt.subplots(figsize = (20,10))
x = cat_info_df.sort_values('avg_review_count', ascending = False).reset_index()['index'][0:20]
y = cat_info_df.sort_values('avg_review_count', ascending = False)['avg_review_count'][0:20]
ax.barh(x,y)
ax.set_title('Top 20 Average Review Counts', fontsize = 15)
ax.set_xlabel('Average Review Count', fontsize = 15)
ax.set_ylabel('Categories', fontsize = 15)
ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
```

```python
fig,ax = plt.subplots(figsize = (20,10))

ax.hist(cat_info_df[cat_info_df['total'] > 5]['avg_review_count'], bins = 8, color = 'b')
avl = plt.axvline(cat_info_df[cat_info_df['total'] > 5]['avg_review_count'].mean(), color='k', linestyle='dashed', linewidth=1)
ax.set_title('Distribution of Average Review Counts by Category', fontsize = 15)
ax.set_xlabel('Average Review Count', fontsize = 15)
ax.set_ylabel('Number of Categories', fontsize = 15)
plt.legend([avl], ['Mean = 222.38'])
plt.show()
```

```python
fig,ax = plt.subplots(figsize = (20,10))

ax.hist(cat_info_df[cat_info_df['total'] > 5]['avg_rating'], bins = 9, color = 'g')
avl = plt.axvline(cat_info_df[cat_info_df['total'] > 5]['avg_rating'].mean(), color='k', linestyle='dashed', linewidth=1)
ax.set_title('Distribution of Average Rating by Category', fontsize = 15)
ax.set_xlabel('Average Rating', fontsize = 15)
ax.set_ylabel('Number of Categories', fontsize = 15)
plt.legend([avl], ['Mean = 3.705'])
plt.show()
```
